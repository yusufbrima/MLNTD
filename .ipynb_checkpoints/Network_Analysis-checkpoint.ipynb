```python
import pandas as pd
import numpy as np
import spacy
from spacy import displacy
import networkx as nx
from pyvis.network import Network
import community.community_louvain as community_louvain
import matplotlib.pyplot as plt
import io
import base64
```

```python
fig_dir = './figures/'
```

```python
def export_network(G, output_format='svg', filename='network_visualization'):
    """
    Export network visualization in different formats
    
    Parameters:
    -----------
    G : networkx.Graph
        The network graph to visualize
    output_format : str, optional (default='svg')
        Export format - can be 'svg', 'png', 'html', or 'base64'
    filename : str, optional
        Output filename (without extension)
    
    Returns:
    --------
    If output_format is 'base64', returns base64 encoded image
    Otherwise, saves file to disk
    """
    # Create Pyvis Network visualization
    net = Network(notebook=True, width="100%", height="700px", 
                  bgcolor='#222222', font_color='white', cdn_resources="in_line")
    
    # Calculate node degrees and set node sizes
    node_degree = dict(G.degree)
    nx.set_node_attributes(G, node_degree, 'size')
    
    # Add graph to network
    net.from_nx(G)
    
    # Export methods
    if output_format == 'html':
        # HTML export (original method)
        net.show(f"./figures/{filename}.html")
        print(f"Network visualization saved as {filename}.html")
    
    elif output_format == 'svg':
        # SVG export using Matplotlib
        plt.figure(figsize=(12, 8))
        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_color='lightblue', 
                node_size=[v * 100 for v in dict(G.degree).values()],
                font_size=8, font_color='black')
        plt.savefig(f"./figures/{filename}.svg", format='svg', bbox_inches='tight')
        plt.close()
        print(f"Network visualization saved as {filename}.svg")
    
    elif output_format == 'png':
        # PNG export using Matplotlib
        plt.figure(figsize=(12, 8))
        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_color='lightblue', 
                node_size=[v * 100 for v in dict(G.degree).values()],
                font_size=8, font_color='black')
        plt.savefig(f"./figures/{filename}.png", format='png', bbox_inches='tight')
        plt.close()
        print(f"Network visualization saved as {filename}.png")
    
    elif output_format == 'base64':
        # Base64 encoded image (useful for embedding in web/jupyter)
        plt.figure(figsize=(12, 8))
        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_color='lightblue', 
                node_size=[v * 100 for v in dict(G.degree).values()],
                font_size=8, font_color='black')
        
        # Save to buffer
        buf = io.BytesIO()
        plt.savefig(buf, format='png', bbox_inches='tight')
        plt.close()
        buf.seek(0)
        
        # Encode to base64
        return base64.b64encode(buf.getvalue()).decode('utf-8')
    
    else:
        raise ValueError("Unsupported output format. Choose 'html', 'svg', 'png', or 'base64'.")


def export_community_network(G, output_format='html', filename='affiliation_communities'):
    """
    Export community network visualization with centrality measures
    
    Parameters:
    -----------
    G : networkx.Graph
        The network graph to visualize
    output_format : str, optional (default='html')
        Export format - can be 'html', 'svg', 'png', or 'base64'
    filename : str, optional
        Output filename (without extension)
    
    Returns:
    --------
    If output_format is 'base64', returns base64 encoded image
    Otherwise, saves file to disk
    """
    # Calculate centrality measures
    degree_dict = nx.degree_centrality(G)
    betweenness_dict = nx.betweenness_centrality(G)
    closeness_dict = nx.closeness_centrality(G)
    
    # Set node attributes
    nx.set_node_attributes(G, degree_dict, 'degree_centrality')
    nx.set_node_attributes(G, betweenness_dict, 'betweenness_centrality')
    nx.set_node_attributes(G, closeness_dict, 'closeness_centrality')
    
    # Detect communities
    communities = community_louvain.best_partition(G)
    nx.set_node_attributes(G, communities, 'group')
    
    # Export methods
    if output_format == 'html':
        # Interactive HTML export (original method)
        com_net = Network(notebook=True, width="100%", height="700px", 
                          bgcolor='#222222', font_color='white', cdn_resources="in_line")
        com_net.from_nx(G)
        com_net.show(f"./figures/{filename}.html")
        print(f"Community network visualization saved as {filename}.html")
    
    elif output_format in ['svg', 'png', 'base64']:
        # Static visualization using Matplotlib
        plt.figure(figsize=(16, 12))
        
        # Use community detection for coloring
        community_colors = plt.cm.rainbow(np.linspace(0, 1, len(set(communities.values()))))
        node_colors = [community_colors[communities[node]] for node in G.nodes()]
        
        # Node size based on degree centrality
        node_sizes = [5000 * degree_dict[node] for node in G.nodes()]
        
        # Create layout
        pos = nx.spring_layout(G, k=0.5, iterations=50)
        
        # Draw network
        nx.draw(G, pos, 
                node_color=node_colors, 
                node_size=node_sizes, 
                with_labels=True, 
                font_size=8, 
                font_color='white',
                edge_color='gray',
                alpha=0.7)
        
        plt.title("Community Network Visualization")
        plt.axis('off')
        
        # Save based on format
        if output_format == 'svg':
            plt.savefig(f"./figures/{filename}.svg", format='svg', bbox_inches='tight', dpi=300)
            print(f"Community network visualization saved as {filename}.svg")
        elif output_format == 'png':
            plt.savefig(f"./figures/{filename}.png", format='png', bbox_inches='tight', dpi=300)
            print(f"Community network visualization saved as {filename}.png")
        elif output_format == 'base64':
            # Save to buffer for base64 encoding
            buf = io.BytesIO()
            plt.savefig(buf, format='png', bbox_inches='tight', dpi=300)
            plt.close()
            buf.seek(0)
            return base64.b64encode(buf.getvalue()).decode('utf-8')
        
        plt.close()
    
    else:
        raise ValueError("Unsupported output format. Choose 'html', 'svg', 'png', or 'base64'.")
```

```python
file_path = './data/merged_data_clean_updated.csv'
# file_path = '/net/store/cv/users/ybrima/mlntd/data/merged_data_clean_final.csv'

df = pd.read_csv(file_path)

df.shape 
```

```python
df.head()
```

```python
# Assuming new_df is already loaded from your previous code
# If you're running this in a new session, load the CSV first:
new_df = df.copy()

# Clean the data - remove NaN, 'nan' strings, and empty strings
new_df = new_df.dropna(subset=['Funding', 'Affiliations','Class'])
new_df = new_df[~new_df['Funding'].str.contains('nan', case=False, na=False)]
new_df = new_df[~new_df['Affiliations'].str.contains('nan', case=False, na=False)]
new_df = new_df[~new_df['Class'].str.contains('nan', case=False, na=False)]

new_df = new_df[new_df['Funding'].str.strip() != '']
new_df = new_df[new_df['Affiliations'].str.strip() != '']
new_df = new_df[new_df['Class'].str.strip() != '']
new_df.head()
```

```python
relationship_df  = new_df[['Funding', 'Affiliations']]


# Sort the cases with a->b and b->a
relationship_df = pd.DataFrame(np.sort(relationship_df.values, axis = 1), columns = relationship_df.columns)
relationship_df

relationship_df["value"] = 1
relationship_df = relationship_df.groupby(["Funding","Affiliations"], sort=False, as_index=False).sum()
relationship_df.head()
```

```python
min_value = 5  # Set your desired minimum value
relationship_df = relationship_df.loc[relationship_df["value"] >= min_value]
print(relationship_df.shape)
relationship_df.head()
```

# Graph analysis and visualization


```python
# Create a graph from a pandas dataframe
G = nx.from_pandas_edgelist(relationship_df, 
                            source = "Funding", 
                            target = "Affiliations", 
                            edge_attr = "value", 
                            create_using = nx.Graph())
```

# Graph visualization - Pyvis


```python
net = Network(notebook = True, width="100%", height="700px", bgcolor='#222222', font_color='white',cdn_resources="in_line")

node_degree = dict(G.degree)

#Setting up node size attribute
nx.set_node_attributes(G, node_degree, 'size')

net.from_nx(G)
net.show("./figures/funding_network.html")
```

# The most important funding flow


```python
# Degree centrality
degree_dict = nx.degree_centrality(G)
# degree_dict
```

```python
degree_df = pd.DataFrame.from_dict(degree_dict, orient='index', columns=['centrality'])
# Plot top 10 nodes
degree_df.sort_values('centrality', ascending=False)[0:9].plot(kind="bar")
plt.show()
```

```python
# Betweenness centrality
betweenness_dict = nx.betweenness_centrality(G)
betweenness_df = pd.DataFrame.from_dict(betweenness_dict, orient='index', columns=['centrality'])
# Plot top 10 nodes
betweenness_df.sort_values('centrality', ascending=False)[0:9].plot(kind="bar")
plt.show()
```

```python
# Closeness centrality
closeness_dict = nx.closeness_centrality(G)
closeness_df = pd.DataFrame.from_dict(closeness_dict, orient='index', columns=['centrality'])
# Plot top 10 nodes
closeness_df.sort_values('centrality', ascending=False)[0:9].plot(kind="bar")
plt.show()
```

```python
# Save centrality measures
nx.set_node_attributes(G, degree_dict, 'degree_centrality')
nx.set_node_attributes(G, betweenness_dict, 'betweenness_centrality')
nx.set_node_attributes(G, closeness_dict, 'closeness_centrality')
```
